#  Movie Revenue & Success Prediction

This project aims to predict a movie's potential box office revenue (regression) and determine whether it will be successful or not (classification), using data from the TMDB 5000 Movie Dataset.

## Objectives:
- Perform exploratory data analysis (EDA)
- Engineer features from text and categorical columns
- Train multiple regression and classification models
- Build pipelines for preprocessing and training
- Evaluate models using appropriate metrics



```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")
```

```python
movies_df = pd.read_csv("../data/tmdb_5000_movies.csv")

movies_df.head()
```

```python
print(f"The dataset contains {movies_df.shape[0]} rows and {movies_df.shape[1]} columns.")
```

```python
movies_df.info()
```

```python
movies_df.isnull().sum().sort_values(ascending=False)
```

```python
movies_df.describe()
```

```python
movies_df.drop("homepage", axis=1, inplace=True)
movies_df.dropna(subset=["release_date"], inplace=True)

mean_runtime = movies_df["runtime"].mean()
movies_df["runtime"] = movies_df["runtime"].fillna(mean_runtime)


zero_budget = (movies_df["budget"] == 0).sum()
zero_revenue = (movies_df["revenue"] == 0).sum()

print(f"Number of movies with zero budget: {zero_budget}")
print(f"Number of movies with zero revenue: {zero_revenue}")
```

## Feature Selection: Dropping Irrelevant or Redundant Columns

In this step, we assess each column to determine whether it is useful for our machine learning models.

We drop the following columns for the reasons below:

- `id`: Just an identifier, not predictive.
- `original_title`, `title`: Duplicate and non-informative for modeling.
- `spoken_languages`, `status`, `tagline`, `homepage`: Mostly redundant or missing values.
- `keywords`, `production_companies`, `production_countries`: Complex textual data or too diverse for this stage.
- `overview`: We will consider it later in NLP-based improvements.

This process leaves us with a cleaner dataset that focuses only on features useful for prediction.


```python
columns_to_drop = [
    "id", "original_title", "spoken_languages", "status",
    "tagline", "title", "keywords", "production_companies",
    "production_countries"
]

movies_df.drop(columns_to_drop, axis=1, inplace=True)

print("Updated DataFrame shape:", movies_df.shape)
movies_df.columns
```

##  Feature Engineering: `release_date` and `genres`

In this step, we transform the `release_date` column to extract the release year, which is more useful for analysis and modeling.

We also process the `genres` column, which is originally a stringified list of dictionaries, to extract the genre names.

This allows us to analyze genre distributions and optionally perform one-hot encoding later on.


```python
movies_df['release_date'] = pd.to_datetime(movies_df['release_date'], errors='coerce')

 
movies_df['release_year'] = movies_df['release_date'].dt.year

 
print(f"Year range: {movies_df['release_year'].min()} - {movies_df['release_year'].max()}")
```

```python
import ast

def extract_genre_names(genre_str):
    try:
        genres = ast.literal_eval(genre_str)
        return [genre['name'] for genre in genres]

    except:
        return []


movies_df['genre_names'] = movies_df['genres'].apply(extract_genre_names)

movies_df[['genres', 'genre_names']].head()
```

## Genre Distribution and One-Hot Encoding

Now that we have extracted genres, we will visualize the most common genres across the dataset.

Then, we will one-hot encode the genres to transform this categorical variable into a numerical format suitable for machine learning models.


```python
 
from collections import Counter

all_genres = sum(movies_df['genre_names'], [])
genre_counts = Counter(all_genres)

 
genre_df = pd.DataFrame(genre_counts.items(), columns=['genre', 'count'])
genre_df = genre_df.sort_values(by='count', ascending=False)

 
plt.figure(figsize=(10, 6))
sns.barplot(x='count', y='genre', data=genre_df, palette='viridis')  
plt.title("Top Genres by Count", fontsize=14)
plt.xlabel("Number of Movies")
plt.ylabel("Genre")
plt.tight_layout()
plt.savefig("../images/top_genres_by_count.png")
plt.show()
```

```python
for genre in genre_df['genre']:
    movies_df[f"genre_{genre.lower()}"] = movies_df['genre_names'].apply(lambda x: int(genre in x))
```

##  Cleaning and Encoding: Language, Budget, and Revenue

In this step, we encode the `original_language` column using one-hot encoding to convert it into a numerical format.

We also remove movies with zero values in either `budget` or `revenue`, as these represent incomplete or unrealistic data.

Finally, we drop textual columns that are no longer needed for modeling.


```python
language_dummies = pd.get_dummies(movies_df['original_language'], prefix='lang')

movies_df = pd.concat([movies_df, language_dummies], axis=1)
```

```python
movie_df = movies_df[(movies_df['budget'] > 0) & (movies_df['revenue'] > 0)]

print(f"Remaining rows after cleaning: {movies_df.shape[0]}")
```

```python
columns_to_drop = ['overview', 'release_date', 'genres', 'genre_names', 'original_language']
movies_df.drop(columns=columns_to_drop, axis=1, inplace=True)

movies_df.head()
```

## Defining Target Variables

We now define the target variables for our two prediction problems:

- **Regression**: Predicting the `revenue` of a movie based on its features.
- **Classification**: Predicting whether a movie is "successful" or not.

We consider a movie to be successful if its `revenue` is at least double its `budget`.
This will serve as our binary classification target.


```python
movies_df['success'] = (movies_df['revenue'] >= 2 * movies_df['budget']).astype(int)

movies_df['success'].value_counts(normalize=True)
```

##  Train-Test Split for Regression and Classification

We now separate the dataset into features (`X`) and targets (`y`) for both regression and classification tasks.

We will use `train_test_split` from scikit-learn to divide the data into training and testing sets.


```python
from sklearn.model_selection import train_test_split

X = movies_df.drop(columns=['revenue', 'success'])

y_reg = movies_df['revenue']
y_clf = movies_df['success']

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X, y_reg, test_size=0.2, random_state=42)

X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X, y_clf, test_size=0.2, random_state=42)

print("Training set shape:", X_train_reg.shape)
print("Test set shape:", X_test_reg.shape)
```
